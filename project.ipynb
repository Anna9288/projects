```python
#get dummy is an encoding inside pandas
#label and one hot encoding is inside machine learning
#one hot encoding drawback is that if the fearures increases the machine takes so much time to learn the data
#get dummy is similar to one hot encoding,but features can be decreased
#in get dummy,first feature will be cut off and using other feature the object will be identified
```

In the input_labels,there are string data which should be converted into numerical data in order to build the model.Here for encoding we use get dummies as it can decrease the features

```python
dummy=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
dummy
```

Here we do encoding in another dataframe and then we join the training dataframe and encoded dataframe into one dataframe

```python
df1=pd.concat([df,dummy],axis=1)
df1
```

Now we drop the unnecessary input labels  that doesn't influence output labels in the new dataframe formed by concatination

```python
df1.drop(['Location','Fuel_Type','Transmission','Owner_Type','Unnamed: 0','Name','New_Price'],axis=1,inplace=True)
df1
```

Now we check the missing values in the dataframe

```python
print(df1.isna().sum())
```

Here in the dataset certain input labels have strings along with numerical data so we replace the string,for that we use str.replace.

```python

df1['Mileage']=df1['Mileage'].str.replace('km/kg','')
df1['Mileage']=df1['Mileage'].str.replace('kmpl','')
df1['Engine']=df1['Engine'].str.replace('CC','')
df1['Power']=df1['Power'].str.replace('bhp','')
df1['Mileage']=df1['Mileage'].str.replace('null','0')
df1['Power']=df1['Power'].str.replace('null','0')
df1['Engine']=df1['Engine'].str.replace('null','0')
df1
```

All  data types should be integer or float type ,we we convert object datatypes into float type.

```python
df1.dtypes 
```

```python

df1['Mileage']=df1['Mileage'].astype(float)
df1['Engine']=df1['Engine'].astype(float)
df1['Power']=df1['Power'].astype(float)
df1.dtypes
```

Now we convert the zeros in respective columns into NaN

```python

df1.loc[df1.Engine==0,'Engine']=np.NaN
df1.loc[df1.Mileage==0,'Mileage']=np.NaN
df1.loc[df1.Power==0,'Power']=np.NaN
print(df1.isna().sum())
```

Now the missing values increased after converting into NaN. Now we fill the missing values with mean and mode.Here we used mode for seats as it cannot be decimal.

```python
df1['Mileage']=df1['Mileage'].fillna(df1['Mileage'].mean())
df1['Power']=df1['Power'].fillna(df1['Power'].mean())
df1['Engine']=df1['Engine'].fillna(df1['Engine'].mean())
df1['Seats']=df1['Seats'].fillna(df1['Seats'].mode()[0])
print(df1.isna().sum())
```

In the dataset, price is the output label and we assign the input labels into x and output label into y.

```python
x=df1.drop(['Price'],axis=1).iloc[:,:]
y=df1.iloc[:,6]
x
```
